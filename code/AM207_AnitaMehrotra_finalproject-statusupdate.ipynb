##### PROBABILISTIC GRAPHICAL MODELS

_REVIEW_

A __Graphical Model__ (GM) is a way to represent a joint distribution by making "conditional independence" (CI) assumptions. They provide a way to define distributions on general, arbitrarily large collections of variables. A "directed graphical model" is a GM whose graph is a __Directed Acyclic Graph__ (DAG). These types of graphical models are more commonly known as __Bayesian networks__, the key stipulation being that parent nodes must come before children nodes and that, of course, there are no undirected cyclic components in the graph structure.

_INFERENCE AND LEARNING_ 

Probabilistic inference in this context means computing $p(D_h|D_v, \theta)$ where $v$ are the visible nodes, $h$ are the hidden nodes, $\theta$ are the known parameters of the model and $D$ is the data. Learning usually means computing a MAP estimate of the parameters given the data, i.e. $D_{i,v}$ are the visible variables in case $i$. 

But as Murphy notes [3], if we adopt a Bayesian view, the parameters are unknown variables and should also be inferred. That is, to a Bayesian, there is no distinction between inference and learning. We can use Expectation-Maximization to infer the hidden parameters.

This goes hand-in-hand with Conditional Distribution Tables. 

##### APPLICATION TO MY PROJECT

In [1], Chetty et al, specify two types of mobility per commuting zone: "relative" and "absolute." They plot a binned scatter plot of a child's income as a rank out of the national incomes versus their parent's income as a rank out of the national incomes of parents. The children were all born between 1980 and 1982. (Studies have shown that income stability occurs at least 20 years after birth.) The plot is binned because of the size of the data: over 40 million children are studied.

<img src="files/rank-rank.png">

In order to obtain the commuting zone mobility, Chetty et al run an OLS on child rank (dependent variable) and parent rank (input feature). Formally, this model can be written as:

$y_{ic} = \alpha_c + \beta_c x_{ic} + \epsilon_{ic}$,

where $y_{ic}$ is child $i$'s national income rank in commuting zone $c$, $x_{ic}$ is the parent of child $i$'s national income rank in commuting zone $c$. We are interested in estimating $\alpha$ and $\beta$.

Once obtaining the estimates, call them $\hat{\alpha_c}, \hat{\beta_c}$, the "relative" mobility is defined as $\hat{\beta_c}$. "Absolute" mobility, denoted $\hat{y_{pc}}$ below, is computed by applying the following function:

$\hat{y_{pc}} = \hat{\alpha_c} + \hat{\beta_c} \cdot p$, 

where $p$ is the parent's national income rank.

This formulation is a 1-dimensional sequence that can be represented as a, albeit simple, DAG:

To infer the parameters of this model, let's run a Bayesian linear regression. We begin by assuming conjugate priors on the parameters of interest and suppose the data points and error term are distributed normally i.i.d.:

$\bar{y} = \bar{\beta} \cdot X + \bar{\epsilon}$, 

where $\bar{y} = 
\left[ \begin{array}{ccc}
y_{0} \\
...\\
y_{n} \end{array} \right]$ with $y_i$ = child income, 

$X = 
\left[ \begin{array}{ccc}
1 & x_{0} \\
1 & x_{1} \\
...\\
1 & x_{n} \end{array} \right]$ with $x_i$ = parent income per county, 

$\bar{\beta} = \left[ \begin{array}{ccc}
\beta_{0} \\
\beta_{1} \end{array} \right]$ such that $\bar{\beta} \sim N(0, \Sigma^{-1})$ (notice $\beta_0$ is an intercept term) and

$\bar{\epsilon}$ is a vector of $\epsilon_i \sim N(0, \sigma^2)$.

##### POSTERIOR

Now the nice thing about this set-up is that we know analytically what the posterior for $\bar{\beta}$ is:

$P(\bar{\beta} | \bar{y}; X, \sigma^2) \propto exp \left( \frac{-1}{2 \sigma^2} \cdot ||\bar{y} - X \bar{\beta}||^2  - \frac{1}{2}\bar{\beta}^T \Sigma \bar{\beta}\right)$,

i.e. it's a multivariate Gaussian by choice of the conjugate prior.

##### MAP

This is nice because now there exists an analytical solution for the parameter estimate of $\bar{\beta}$:

$\bar{\hat\beta} \sim N(\mu_n, \Sigma_n)$ where

$\mu_n = (X^TX + \sigma^2 \Sigma)^{-1}X^T\bar{y}$ and $\Sigma_n = \sigma^2(X^TX + \sigma^2\Sigma)^{-1}$.

```python
# define functions for hyper-parameters
def mu(X, sigma, Sigma, y):
    
    first_term = np.dot(X.T, X)
    second_term = (sigma**2) * Sigma
    the_sum_inv = np.linalg.inv(first_term + second_term) # most computationally intensive calculation 
    
    return np.dot(the_sum_inv, np.dot(X.T, y)) 

def covariance(sigma, X, Sigma):
    
    first_term = np.dot(X.T, X)
    second_term = (sigma**2) * Sigma
    the_sum_inv = np.linalg.inv(first_term + second_term)
    
    return (sigma**2) * the_sum_inv
```

```python
# initialize variables

# create y vector
y = (np.log(child)).reshape(len(child), 1)

# create X matrix
first_Xcol = np.ones(len(parent)).reshape(len(parent), 1)
X = np.hstack((first_Xcol, (np.log(parent)).reshape(len(parent),1)))

# create Sigma matrix
Sigma = np.identity(2)

# create sigma
sigma = np.random.rand() ############## 1? does it matter? ##############

# compute hyper-parameters

print "mu:"
my_mu = mu(X, sigma, Sigma, y)
print my_mu
print
print "covariance:"
my_cov = covariance(sigma, X, Sigma)
print my_cov

# derive beta estimate
print
print "estimated beta:"
mean_vec = [my_mu[0][0], my_mu[1][0]]
beta = np.random.multivariate_normal(mean_vec, my_cov, 1)[0]
print beta
```

```python
xvals = np.linspace(10,14)
epsilon = np.random.normal(loc=0, scale=sigma, size=len(xvals))

the_fxn = lambda x: beta[0] + x*beta[1] + epsilon
yvals = the_fxn(xvals)

the_fxn2 = lambda x: beta[0] + x*beta[1]
yvals2 = the_fxn2(xvals)

plt.figure(figsize=[10,8])
pplt.plot(xvals, yvals, "--", color="blue")
pplt.plot(xvals, yvals2, color="brown", linewidth=1.5)
pplt.scatter(np.log(parent), np.log(child), color="orange", alpha=0.2)
xlim([10,13.25])
plt.grid(True)
plt.title("Line of Best Fit with Estimated Parameters")
plt.xlabel(r"$\log(parent)$")
plt.ylabel(r"$\log(child)$")
plt.legend(["with noise","without noise","original data"])

plt.show()
```
