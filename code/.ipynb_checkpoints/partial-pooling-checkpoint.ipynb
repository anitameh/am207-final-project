```python
# now standardize all of the data
some_features = pd.DataFrame(all_data, columns = ["State", "stfips", "cntyfips", "Mean Child Income", "Mean Parent Income", 
                                                  "School Expenditure per Student", "Teacher Student Ratio", "Test Score Percentile", 
                                                  "High School Dropout Rate", "Teenage Birth Rate"])

M,N = shape(some_features)[0], shape(some_features)[1]
col_names = ["Mean Child Income", "Mean Parent Income", "School Expenditure per Student",
            "Teacher Student Ratio", "Test Score Percentile", "High School Dropout Rate",
            "Teenage Birth Rate"]
index = np.arange(0, 2769)
standardized_data = pd.DataFrame(index=index, columns=col_names)
standardized_data["State"] = all_data[["State"]]
standardized_data["stfips"] = all_data[["State FIP"]]
standardized_data["cntyfips"] = all_data[["County ID"]]

population = all_data[["Number of Children in Core Sample"]].values
sqrt_pop = np.sqrt(population)

for i in xrange(len(col_names)):
    one_feature =  some_features[[ col_names[i] ]].values
    one_mean = np.nanmean(one_feature)
    
    # subtract mean
    demeaned = one_feature - one_mean
    
    # 'divide' by SD
    std_col = demeaned*sqrt_pop
    
    standardized_data[ col_names[i] ] = std_col
```

Now let's compute the state-level predictors. We want this for all features, even though in the first partial pooling iteration, we will focus on the mean parent income.

First, let's try this on a toy example:

```python
index = np.arange(0,10)
col_names = ["mpi", "ctyfip", "stfip"]
testdf = pd.DataFrame(index=index, columns=col_names)

testdf["mpi"] = standardized_data[["Mean Parent Income"]].values[0:10]
testdf["stfip"] = np.array([1,1,1,1,1,2,2,2,3,3])
testdf["ctyfip"] = np.array([10,11,12,13,14,20,21,22,30,31])

testdf
```

```python
unique_states = np.unique(testdf.stfip)
list_of_aves = []

for us in unique_states:
    # extract mean parent income for state id = us
    sub = testdf[testdf.stfip == us].mpi
    
    # compute mean for that state
    state_ave = np.nanmean(sub)
    num_counties = len(sub)
    
    # construct new column
    list_of_aves.append( list(state_ave*np.ones(num_counties)) )
    
flattened = [item for sublist in list_of_aves for item in sublist]
# print flattened

testdf["state mpi"] = flattened
testdf
```

Great! Now that we know the above works, let's apply it to the entire data set for mean parent income:

```python
def get_state_aves(df, col_name, new_col_name):
    
    unique_states = np.unique(df.stfips)
    list_of_aves = []
    
    for us in unique_states:
        # extract county-level mean parent income for state id = us
        sub_df = df[df.stfips == us]
        sub = sub_df[col_name]
        
        # compute mean for that state
        ave = np.nanmean(sub)
        num_counties = len(sub)
        
        # construct new column
        list_of_aves.append( list(ave*np.ones(num_counties)) )
        
    flattened = [item for sublist in list_of_aves for item in sublist]
    df[new_col_name] = flattened

    return df
```

```python
new_df = get_state_aves(standardized_data, "Mean Parent Income", "State MPI")
new_df.head()
```
