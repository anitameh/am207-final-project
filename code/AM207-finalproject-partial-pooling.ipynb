First, we must standardize the data!

$\bar{y_i} = \frac{y_i - \mu_{y}}{\sigma_{y}} \Rightarrow \bar{y_i} \propto (y_i - \mu_y) \cdot \sqrt{population_i}$

In this first case, we use a hierarchical model to understand the state effect on mean child income per county. This is the first step (next, we will include back in the regression). The hiearchical model looks like this:

<img src="files/state-effect.png">

Let $y_{ij}$ = the mean child income in county $i$ and state $j$. Then $\theta_j$ captures the state effect. Furthermore, assume

$y_{ij}|\theta_j \sim N(\theta_j, \sigma^2)$, with

$i = {1, ..., n_j}$ where $n_j$ = the number of observations in state $j$, and $j = 1, ..., J$.

Assume that $\theta_j$ are samples from a Normal distribution with hyperparameters $\phi = \{\mu, \tau\}$. 

##### LIKELIHOOD

We can thus write down the likelihood:

$p(\theta_1, ... \theta_J | \mu , \tau) = \prod \limits_{j=1}^J N(\theta_j | \mu, \tau^2)$.

Then this means $p(\theta_1, ... \theta_J) = \int \prod \limits_{j=1}^J [N(\theta_j|\mu, \tau^2)] \cdot p(\mu, \tau) d(\mu, \tau)$.

```python
# compute sample mean, sample variance
unique_sample_mean = []
unique_sample_var = []
sample_mean = []
sample_var = []
unique_states = np.unique(std_data.state)
var = np.var(std_mci)
print "population variance =", var
print

for us in unique_states:
    
    # extract for state id = us
    sub = std_data[std_data.state == us].std_mci
    
    # mean child income per state
    state_ave = np.mean(sub)
    
    # mean variance per state
    num_counties = len(sub)
    state_var = var/num_counties
    
    # store
    unique_sample_mean.append(state_ave)
    unique_sample_var.append(state_var)
    
    sample_mean.append( list(state_ave*np.ones(num_counties)) )
    sample_var.append( list(state_var*np.ones(num_counties)) )

sample_means = [item for sublist in sample_mean for item in sublist]  
sample_vars = [item for sublist in sample_var for item in sublist]  
print "sample means, head"
print sample_means[0:10]
print
print "sample variances, head"
print sample_vars[0:10]
```

```python
# compute weighted average
inv_var = 1./np.array(sample_vars)
prod = inv_var * np.array(sample_means)

numerator = sum(prod)
denominator = sum(inv_var)

weighted_ave = numerator/denominator
print "weighted average =", weighted_ave
```

Now suppose non-informative priors for the hyperparameters, where the prior for $\mu$ is in fact uniform ($p(\mu) \propto 1$). Then 

$p(\mu, \tau) = p(\mu| \tau) \cdot p(\tau) \propto p(\tau)$.

##### JOINT POSTERIOR

Now we can write down the joint posterior, up to a normalizing constant:

$p(\theta, \mu, \tau | y) \propto p(\mu, \tau) \cdot p(\theta | \mu, \tau) \cdot p(y | \theta) = p(\mu, \tau) \cdot \prod \limits_j N(\theta_j | \mu, \tau^2) \cdot \prod \limits_j N(\bar{y_{.j}} | \theta_j, \sigma_j^2)$,

where $\bar{y_{.j}}$ is the sample mean for state $j$ and $\sigma_j$ is the sample variance for state $j$, both computed above.

##### CONDITIONAL POSTERIOR

So NOW, we can write down the distributions for each of the $\theta_j$'s (they are independent and normal):

$\theta_j | \mu, \tau, y \sim N(\hat{\theta_j}, V_j)$, where

$\hat{\theta_j} = \frac{(1/\sigma_j^2) \bar{y_{.j}} + (1/\tau^2)\mu}{(1/\sigma_j^2) + (1\tau^2)}$ and $V_j = \frac{1}{(1/\sigma_j^2) + (1/\tau^2)}$.

```python
# compute \hat{\theta_j}
def theta_hat(j, mu, tau):
    
    mean_j = sample_means[j]
    var_j = sample_vars[j]
    
    # numerator
    first_term = (1./var_j)*mean_j
    second_term = (1./tau**2)*mu
    numerator = first_term + second_term
    
    # denominator
    denominator = (1./var_j) + (1./(tau**2))
    
    return numerator / denominator

# compute V_j
def V(j, tau):
    
    var_j = sample_vars[j]
    
    denominator = (1./var_j) + (1./(tau**2))
    
    return 1./denominator
```

##### MARGINAL POSTERIOR DISTRIBUTIONS (FOR THE HYPERPARAMETERS)

For the hyperparameters, we have the following:

$p(\mu, \tau | y) \propto p(\mu, \tau) \cdot p(y | \mu, \tau) = p(\mu, \tau) \cdot \prod \limits_{j=1}^J N(\bar{y_{.j}} | \mu, \sigma_j^2 + \tau^2)$.

So now we can find the posterior distribution of $\mu$ condtional on $\tau$ by factorizing:

$p(\mu, \tau | y) = p(\mu | \tau, y) \cdot p(\tau, y)$. 

With a uniform prior for $p(\mu | \tau)$, we have $\mu | \tau, y \sim N(\hat{\mu}, V_{\mu})$,

where $\hat{\mu} = \frac{\sum \limits_j \frac{1}{\sigma_j^2 + \tau^2} \bar{y_{.j}} }{\sum \limits_j \frac{1}{\sigma_j^2 + \tau^2}}$ and $V_{\mu}^{-1} = \sum \limits_j \frac{1}{\sigma_j^2 + \tau^2}$.

Assume a uniform prior on $\tau$: $p(\tau) \propto 1$. This produces a proper posterior prior.

```python
# compute inv_V_mu
def inv_V_mu(tau):
    
    v = np.array(sample_vars)
    
    return sum( 1./(v + tau**2) )
```

```python
tau = np.arange(0,5000)/50.
tau = tau[4999]
inv_V_mu(tau)
```

```python
# compute \hat{\mu}
def mu_hat(tau):
    
    y = np.array(sample_means)
    v = np.array(sample_vars)
    
    # numerator
    inside = (1./(v + tau**2)) * y
    numerator = sum(inside)
    
    # denominator
    denominator = inv_V_mu(tau)
    
    return numerator / denominator
```

```python
tau = np.arange(0,5000)/50.
tau = tau[4999]
mu_hat(tau)
```

```python
# compute P(tau | y)
def p_tau_y(tau, vmu, muhat):
    
    y = std_mci
    v = np.array(sample_vars)
    
    return np.sqrt( vmu * np.prod( 1./(tau**2 + v) )) * np.prod( np.exp(-0.5*(y - muhat)*(y - muhat)/(tau**2 + v )) )
    
```

##### POSTERIOR SIMULATIONS

Okay, so we have this:

$p(\theta, \mu, \tau | y)$  

$\propto p(\mu, \tau) \cdot p(\theta | \mu, \tau) \cdot p(y | \theta)$ 

$= p(\tau | y) \cdot p(\mu | \tau, y) \cdot p(\theta| \mu, \tau, y) \cdot p(y | \theta)$.

 * __Simulations for $\tau$:__ This can be done from a uniform grid in the $\tau$ values, with the function $p(\tau|y)$.
 * __Simulations for $\mu, \theta$:__ These come from the corresponding normals.
 * __Likelihood__: Recall that the final term $p(y | \theta)$ is simply the likelihood.
